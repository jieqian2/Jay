# PART II
Preprocessing, feature extraction, feature selection

```python
X=df.iloc[:,:-3]
y=df.iloc[:,-2]
y_mul=df.iloc[:,-1]
normalized_X = X.apply(lambda x : (x-np.min(x))/(np.max(x)-np.min(x)))
normalized_X.describe()
```

# 
Feature Extraction - PCA

```python
# Reference Datacamp
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

# Create scaler: scaler
scaler = StandardScaler()

# Create a PCA instance: pca
pca = PCA()

# Create pipeline: pipeline
pipeline = make_pipeline(scaler,pca)

# Fit the pipeline to 'X'
pipeline.fit(X)

# Plot the explained variances
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_)
plt.xlabel('PCA feature')
plt.ylabel('variance')
plt.title("PCA Feature")
plt.show()
```

```python
pca=PCA(n_components=7)
pca.fit(normalized_X)
pca_X=pca.transform(normalized_X)
print(pca_X.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(pca_X,y,test_size=0.2,random_state=42)
```

#  PART III Binary Class
Model fitting and evaluation-- Overview

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error

# Set seed for reproducibility
SEED=42
# Instantiate lr
lr = LogisticRegression(random_state=SEED)
# Instantiate knn
knn = KNN(n_neighbors=10)
# Instantiate dt
dt = DecisionTreeClassifier(min_samples_leaf=0.13, random_state=SEED)
# Instantiate SVC
svc=SVC(kernel="rbf")
# Define the list classifiers
classifiers = [('Logistic Regression', lr), ('K Nearest Neighbours', knn), ('Classification Tree', dt),('SVC',svc)]

# Iterate over the pre-defined list of classifiers
for clf_name, clf in classifiers:     
    # Fit clf to the training set
    clf.fit(X_train, y_train)       
    # Predict y_pred
    y_pred = clf.predict(X_test)   
    # Calculate accuracy
    accuracy = accuracy_score(y_pred, y_test) 
    # Evaluate clf's accuracy on the test set
    print('{:s} : {:.3f}'.format(clf_name, accuracy)) 
    # Evalutate MSE
    print('MSE test: %.3f' % (mean_squared_error(y_test, y_pred)))
    # Evalutate R^2
    # print('R^2 test: %.3f' % (r2_score(y_test, y_pred)))
    print("-------------------------------------------")
    
```

# 
voting classifier

```python
from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
# Set seed for reproducibility
SEED=42
# Instantiate lr
lr = LogisticRegression(C=13,solver='lbfgs',random_state=SEED)
# Instantiate knn
knn = KNN(n_neighbors=10)
# Instantiate dt
dt = DecisionTreeClassifier(criterion='gini',max_depth=2,max_features=2,min_samples_leaf=8, random_state=SEED)
# Instantiate SVC
svc=SVC(kernel="rbf",C=14,gamma=5)
# Define the list classifiers
classifiers = [('Logistic Regression', lr), ('K Nearest Neighbours', knn), ('Classification Tree', dt),('SVC',svc)]

# Iterate over the pre-defined list of classifiers
for clf_name, clf in classifiers:     
    # Fit clf to the training set
    clf.fit(X_train, y_train)       
    # Predict y_pred
    y_pred = clf.predict(X_test)   
    # Calculate accuracy
    accuracy = accuracy_score(y_pred, y_test) 
    # Evaluate clf's accuracy on the test set
    print('{:s} : {:.3f}'.format(clf_name, accuracy)) 
    # Evalutate MSE
    # print('MSE test: %.3f' % (mean_squared_error(y_test, y_pred)))

vc=VotingClassifier(estimators=classifiers)

vc.fit(X_train,y_train)
y_pred=vc.predict(X_test)

print('Voting Classifier:{}'.format(accuracy_score(y_test,y_pred)))
```

# 
Adaboost

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import AdaBoostClassifier

# Instantiate dt with the best
best_dt = DecisionTreeClassifier(max_depth=2)

param_dist={'n_estimators':[120,140,160,180,200]}

# Instantiate ada
ada = AdaBoostClassifier(base_estimator=best_dt, random_state=SEED)
gs_cv = GridSearchCV(ada, param_dist, cv=10)

# Fit ada to the training set
gs_cv.fit(X_train, y_train)
print("Adaboost Decision Tree Parameters: {}".format(gs_cv.best_params_))
print("Best score is {}".format(gs_cv.best_score_))
```

```python
best_ada = AdaBoostClassifier(base_estimator=best_dt, n_estimators=200, random_state=SEED)
best_ada.fit(X_train,y_train)
y_pred=best_ada.predict(X_test)
accuracy = accuracy_score(y_pred, y_test) 
print('{:s} : {:.3f}'.format("Adaboost Accuracy", accuracy)) 
```

```python
# Import roc_auc_score
from sklearn.metrics import roc_auc_score
ada = AdaBoostClassifier(base_estimator=best_dt, n_estimators=200, random_state=SEED)
ada.fit(X_train, y_train)
# Compute the probabilities of obtaining the positive class
y_pred_proba = ada.predict_proba(X_test)[:, 1]

# Evaluate test-set roc_auc_score
ada_roc_auc = roc_auc_score(y_test, y_pred_proba)

# Print roc_auc_score
print('ROC AUC score: {:.2f}'.format(ada_roc_auc))
```

```python
from sklearn.metrics import roc_curve
fpr,tpr,thresholds=roc_curve(y_test,y_pred_proba)
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr,tpr,label='L R')
plt.show()
```
